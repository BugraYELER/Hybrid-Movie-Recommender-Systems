```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")
```

```python
ratings.head()
```

```python
avg_rating = ratings.groupby("movieId").mean()["rating"]
count_rating = ratings.groupby("movieId").count()["rating"]
```

```python
avg_rating.head()
```

```python
count_rating.head()
```

```python
movies = movies.merge(count_rating,left_on=["movieId"],right_on = ["movieId"])
```

```python
movies = movies.rename(columns={'rating':'vote_count'})
```

```python
movies = movies.merge(avg_rating,left_on=["movieId"],right_on = ["movieId"])
movies = movies.rename(columns={'rating':'avg_rating'})
```

```python
movies.head()
```
